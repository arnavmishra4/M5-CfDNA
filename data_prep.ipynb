```python
import json

# Constraint defined in model.py
MAX_FRAGS_PER_REGION = 64

# Load your raw data
with open("testing.json", "r") as f:
    raw_data = json.load(f)

perfect_input = {"regions": {}}

for region_key, region_data in raw_data.get("regions", {}).items():
    # Enforce the exact max fragments filtering rule used by the model
    fragments = region_data.get("fragments", [])[:MAX_FRAGS_PER_REGION]
    if not fragments:
        continue

    # Clean individual fragments to only contain keys the model processes
    filtered_fragments = []
    for frag in fragments:
        clean_frag = {}
        if "tokens" in frag:
            clean_frag["tokens"] = frag["tokens"]
        if "cls_vector" in frag:
            clean_frag["cls_vector"] = frag["cls_vector"]
        
        if clean_frag:
            filtered_fragments.append(clean_frag)

    if not filtered_fragments:
        continue

    # Reconstruct the optimized region entry
    perfect_input["regions"][region_key] = {
        "chromosome": str(region_data["chromosome"]),
        "genomic_start": int(region_data["genomic_start"]),
        "fragments": filtered_fragments
    }

# Save the perfect, production-ready JSON file
with open("perfect_model_input.json", "w") as f:
    json.dump(perfect_input, f, indent=2)

print("Saved optimized input file to 'perfect_model_input.json'")
```
